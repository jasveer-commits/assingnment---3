How it works:

Loop through each number in the list.

Ignore numbers less than or equal to 1.

Check if the number is divisible by any number from 2 to num - 1.

If it is not divisible by any of them, it is prime.

Print the prime numbers.

```python


numbers = [10, 3, 5, 8, 11, 14, 17, 20, 23]

print("Prime numbers in the list:")

for num in numbers:
    # Prime numbers are greater than 1
    if num > 1:
        is_prime = True
        
        # Check divisibility from 2 to num-1
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        
        if is_prime:
            print(num)
```

How it works:

The function analyze_numbers() takes a list as input.

It uses a loop to check each number.

If the number is even (num % 2 == 0), it increases even_count.

Otherwise, it adds the number to odd_sum.

Finally, it returns both values.

```python


def analyze_numbers(numbers):
    even_count = 0
    odd_sum = 0

    for num in numbers:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_sum += num

    return even_count, odd_sum


# Example usage
nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]

even_count, odd_sum = analyze_numbers(nums)

print("Count of even numbers:", even_count)
print("Sum of odd numbers:", odd_sum)
```

Explanation:

principal is required.

rate has a default value of 5%.

time has a default value of 1 year.

If rate and time are not provided, the default values are used.

The function is demonstrated using:

Positional arguments

Default arguments

Keyword arguments

Mixed arguments

```python
# Function to calculate simple interest
def calculate_simple_interest(principal, rate=5, time=1):
    simple_interest = (principal * rate * time) / 100
    return simple_interest


# Calling function using positional arguments
si1 = calculate_simple_interest(1000, 10, 2)
print("Simple Interest (Positional Arguments):", si1)

# Calling function using default arguments (only principal provided)
si2 = calculate_simple_interest(1500)
print("Simple Interest (Using Default Rate and Time):", si2)

# Calling function using keyword arguments
si3 = calculate_simple_interest(principal=2000, rate=8, time=3)
print("Simple Interest (Keyword Arguments):", si3)

# Mixing positional and keyword arguments
si4 = calculate_simple_interest(2500, time=4)
print("Simple Interest (Mixed Arguments):", si4)
```

How it works:

The __init__ method initializes name and marks.

The calculate_grade() method determines the grade based on marks.

Two objects (student1 and student2) are created.

Their grades are calculated and displayed.

```python
# Define Student class
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Method to calculate grade
    def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        elif self.marks >= 50:
            return "D"
        else:
            return "F"


# Create objects
student1 = Student("Alice", 85)
student2 = Student("Bob", 72)

# Display grades
print("Name:", student1.name)
print("Marks:", student1.marks)
print("Grade:", student1.calculate_grade())
print()

print("Name:", student2.name)
print("Marks:", student2.marks)
print("Grade:", student2.calculate_grade())
```

Explanation:

__balance is a private attribute (name mangling prevents direct access).

deposit() adds money if the amount is valid.

withdraw() subtracts money only if sufficient balance exists.

get_balance() allows controlled access to the balance.

Direct access like account.__balance is not allowed.

```python
# Define BankAccount class
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance   # Private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance or invalid amount.")

    # Method to check balance (controlled access)
    def get_balance(self):
        return self.__balance


# Create object
account = BankAccount(1000)

# Perform operations
account.deposit(500)
account.withdraw(300)

# Access balance using method (not directly)
print("Current Balance:", account.get_balance())

# The following would cause an error if uncommented:
# print(account.__balance)  # Cannot access private attribute directly
```

How it works:

ValueError handles non-numeric input (like letters).

ZeroDivisionError handles attempts to divide by zero.

else runs if no exception occurs.

finally always runs, whether an error occurs or not.

```python
# Program to perform division with exception handling

try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    result = num1 / num2
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Invalid input. Please enter numeric values only.")

except Exception as e:
    print("Unexpected error occurred:", e)

else:
    print("Division performed successfully.")

finally:
    print("Program execution completed.")
```

How it works:

InvalidAgeError is a custom exception class derived from Exception.

The check_age() function raises InvalidAgeError if age is less than 0 or greater than 120.

ValueError handles non-integer input.

The custom exception is caught separately and displays a meaningful message.

```python
# Define custom exception
class InvalidAgeError(Exception):
    pass


# Function to check age
def check_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError("Age must be between 0 and 120.")
    else:
        print("Valid age entered:", age)


# Main program
try:
    user_age = int(input("Enter your age: "))
    check_age(user_age)

except ValueError:
    print("Invalid input. Please enter a valid integer.")

except InvalidAgeError as e:
    print("Custom Exception Caught:", e)

finally:
    print("Program finished.")
```

How it works:

open(filename, "w") creates the file (or overwrites if it already exists).

write() adds student records to the file.

open(filename, "r") opens the file for reading.

read() retrieves and displays the content.

with ensures the file is properly closed after operations.

```python
# File name
filename = "students.txt"

# Writing student data to file
with open(filename, "w") as file:
    file.write("Name: Alice, Marks: 85\n")
    file.write("Name: Bob, Marks: 78\n")
    file.write("Name: Charlie, Marks: 92\n")

print("Student data written to file successfully.\n")

# Reading and displaying file content
with open(filename, "r") as file:
    content = file.read()
    print("Reading file content:\n")
    print(content)
```

How it works:

try block attempts to open and read the file.

Each line is stripped of whitespace and converted to a number.

sum() calculates the total, and the average is computed safely.

FileNotFoundError is handled if the file doesn’t exist.

ValueError is handled in case the file contains non-numeric lines.

except Exception catches any unexpected errors.

```python
filename = "numbers.txt"

try:
    with open(filename, "r") as file:
        numbers = []
        for line in file:
            # Remove whitespace and convert to float
            numbers.append(float(line.strip()))

    total = sum(numbers)
    average = total / len(numbers) if numbers else 0

    print("Numbers:", numbers)
    print("Total:", total)
    print("Average:", average)

except FileNotFoundError:
    print(f"Error: The file '{filename}' does not exist.")

except ValueError:
    print("Error: File contains non-numeric values.")

except Exception as e:
    print("Unexpected error:", e)
```
